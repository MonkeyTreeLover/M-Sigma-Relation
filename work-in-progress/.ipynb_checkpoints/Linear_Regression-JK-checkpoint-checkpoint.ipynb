Firstly the correlation coefficient $r=\rho$ were calculated, to know if a correlation can be assumed or not:
$$ r = \frac{\sum{}(x_i-\overline{x})(y_i-\overline{y})}{\sqrt{\sum{}(x_i-\overline{x})^2 \sum{}(y_i-\overline{y})^2}}$$

r can be between -1 and 1. If the R=0, than no relation can be assumed. The greater the absolute value is, the greater the correlation. -1 is for a negativ slope (strong anticorrelation) and +1 for a positv slope (strong correlation).

A t-test on r tells, whether r differs significantly from 0 (for a data set of n data points): 
$$ t= \frac{r\sqrt{n-2}}{\sqrt{1-r^2}} $$

To get the p-value, it is "necessary" to look at at table or calculate it with a computer. 

The calculations were done with scipy.stats.pearsonr

```python
from scipy.stats import pearsonr

x = "sigma_e (km/s)"
y = "M_BH (solar masses)"
x = "sigma_e"
y = "M_BH"
ylog = "Log_Mbh"
xlog = 'log_S'

M_star = 1/1.989e30
sig = 1/200
M_star = 1
sig = 1
nData = {"Early": [np.log(early_type[x]*sig)/np.log(10), np.log(early_type[y]*M_star)/np.log(10)],
        "Late"  : [np.log(late_type[x]*sig)/np.log(10), np.log(late_type[y]*M_star)/np.log(10)],
        "AGN"   : [np.log(data_AGN_SI*sig)/np.log(10), np.log(data_AGN_BH*M_star)/np.log(10)],    
        "Guel_t": [np.log10(data[x]*sig), np.log10(data[y]*M_star)]}

nDataL = {"Early": [early_type[xlog]*sig, early_type[ylog]*M_star],
        "Late"  : [late_type[xlog]*sig, late_type[ylog]*M_star],
        "Guel_t": [data[xlog]*sig, data[ylog]*M_star],
        "AGN"   : [np.log10(data_AGN_SI*sig), np.log10(data_AGN_BH*M_star)]}    

nDataRaw = {"Early": [early_type[x]*sig, early_type[y]*M_star],
        "Late"  : [late_type[x]*sig, late_type[y]*M_star],
        "Guel_t": [data[x]*sig, data[y]*M_star],
        "AGN"   : [np.log10(data_AGN_SI*sig), np.log10(data_AGN_BH*M_star)]}    

r = {}
p_value = {}
rL = {}
p_valueL = {}
#for i, v in enumerate(nData):
for i, v in enumerate(nDataL):
    rr, p = pearsonr(nData[v][0], nData[v][1])
    rrL, pL = pearsonr(nDataL[v][0], nDataL[v][1])
    r[v] = rr
    rL[v] = rrL
    p_value[v] = p
    p_valueL[v] = pL

    temp = "rejected"
    if p_value[v] <= 0.05:
        temp = "not rejected"
    #print(f"{v}: {temp}")
    #print(f"r={r[v]:.3f}")
    #print(f"p={p_value[v]*100:.3f}%\n")

    temp = "not significant"
    if p_valueL[v] <= 0.05:
        temp = "significant - there is a correlation"
    print(f"{v}: {temp}")
    print(f"r={rL[v]:.3f}")
    print(f"p={p_valueL[v]*100:.3f}%\n")

```

With a signifcant level of 95%, only the late types can be rejected. But for the AGN the r value is very small, thefore the probabilty, if there is a realtionship is very low.

In the this section we try a linear regression. 
We use the statsmodels.api libary for this and also for calulating values. 

For the Linear Regression the following equation is used:
$$ \log \left( \frac{M}{M_\odot} \right) = \alpha + \beta \log \left( \frac{\sigma}{200} \right) $$

This equation is based on the same pattern as the approach in the paper from Gültekin et al., which enables us to compare our results with litature. 
For right use of the varaible, it is necessary to remove $\log 200$ from $$\log \sigma$$.  

The value bse of the libaray is the standard Error and can be calculated as followed: 
$$ SE(\hat{\beta_i}) = \sqrt{\hat{\sigma^2}(X^TX)_{ii}^{-1}} $$

$\beta_i$ are the coeffeicents, X is a matrix and $\sigma^2$ is the resedium error. 

```python
import statsmodels.api as sm

####################### LSM - approach JK ######################
x = "sigma_e"
x200 = "s_2ßß"
y = "M_BH" 
ylog = "Log_Mbh"

TabRaw[x200] = np.log10(TabRaw[x]/200)
#print(data[x200])
mean_S = np.mean(TabRaw[x200])
print("Mean log(sigma/200): ", mean_S)
mean_M = np.mean(TabRaw[ylog])
print("Mean log(Mbh): ", mean_M)

#least square: g(x) = b0 + b1*x
sum_MS_dev = 0.0
sum_M_dev_2 = 0.0

for index, row in TabRaw.iterrows():
#    print(row)
    sum_MS_dev +=  (row[x200] - mean_S)*(row[ylog] - mean_M)
    sum_M_dev_2 += (row[x200] - mean_S)**2

b1 = sum_MS_dev/sum_M_dev_2
b0 = mean_M- b1*mean_S

print("Parameters of straight: log(Mbh) = ", b0, " + ", b1," * log(sigma/200)\n")
####################### LSM - approach JK ######################


#fitData_WTF = nData[0:-1][0].copy()
#fitData_WTF = [nDataL[v][0] - np.log(200) for v in nDataL]
#print(np.log10(10))
fitData_WTF = [nData[v][0] - np.log10(200) for v in nData]
results = []

RESULT = []

for i, v in enumerate(nData):
    fitData_WTF[i] = sm.add_constant(fitData_WTF[i])
    #model = sm.OLS(nData[v][1]-np.log(1.989e30)/np.log(10), fitData_WTF[i])
    model = sm.OLS(nData[v][1], fitData_WTF[i])
    results.append(model.fit())
    #print(results[-1].summary())
    #print("\n+++++++++++++++n")
    #print(results[-1].params)

    #temp = results[-1].params[0]+np.log(1.989e30)/np.log(10)#+results[-1].params[1]*np.log(200)/np.log(10)
    temp = results[-1].params[0]
    RESULT.append((results[-1].params[1], temp))
    print(v, ": ", RESULT[-1])
    #print("\n*******************\n")
    
```
