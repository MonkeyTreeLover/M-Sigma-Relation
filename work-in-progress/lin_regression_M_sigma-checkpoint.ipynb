```python
import statistics as stat
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

File = "data/M-sig_gueltekin_expanded.csv"
data = ascii.read(File) 
TabRaw = pd.read_csv(File)

data_by_type = data.group_by("Type")
early_type = data_by_type.groups[0:9]
late_type = data_by_type.groups[9:18]
#print(TabRaw)
```

```python
y = "sigma_e"
x = "M_BH" 
xlog = "Log_Mbh"
ylog = 'log_S'

TabRaw[ylog] = np.log10(TabRaw[y])

mean_M = stat.mean(TabRaw[xlog])
print("Mean log(Mbh): ", mean_M)
mean_S = stat.mean(TabRaw[y])
print("Mean sigma: ", mean_S)
mean_Sl = stat.mean(TabRaw[ylog])
print("Mean log(sigma): ", mean_Sl)

#least square: g(x) = b0 + b1*x
sum_MS_dev = 0.0
sum_M_dev_2 = 0.0
for index, row in TabRaw.iterrows():
#    print(row)
    sum_MS_dev +=  (row[xlog] - mean_M)*(row[y] - mean_S)
    sum_M_dev_2 += (row[xlog] - mean_M)**2

b1 = sum_MS_dev/sum_M_dev_2
b0 = mean_S - b1*mean_M

print("Parameters of straight: sigma = ", b0, " + ", b1," * log(M_bh)\n")
sigStart = b0 + b1*6.4
sigEnd = b0 + b1*9.8
print("START: log(M_bh) = 6.5, ",sigStart)
print("END: log(M_bh) = 9.7, ",sigEnd,"\n")
P_S = [6.4,9.8]
P_E = [sigStart,sigEnd]

plt.scatter(early_type[xlog], early_type[y], label="Early-type")
plt.scatter(late_type[xlog], late_type[y], label="Late-type")
plt.plot(P_S, P_E, color="green")
    
plt.title("log(M)-sigma")
plt.legend()
plt.xlabel("log(M_bh)")
plt.ylabel("sigma")
plt.show()
```

The plot shows a linear regression for the given sample of the M-sigma relation estimated by the least square method.
