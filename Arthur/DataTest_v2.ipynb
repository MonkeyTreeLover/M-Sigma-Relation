```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

file = "star_data.fits"

hdulist  = fits.open(file)

data = hdulist[1].data
```

```python
plt.scatter(data["Log M"], data["sigma"])
    
plt.xscale('log')
plt.yscale('log')

plt.title("M-Sigma")
#plt.legend()
plt.xlabel("Log M")
plt.ylabel("sigma ")



plt.show()
```
